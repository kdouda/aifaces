```python
import cv2
from matplotlib import pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('bbox_celeb.csv')
df.describe()

row = df[:1]
print(row)
```

```python
def extract_face(in_filename : str, out_filename : str, x : int, y : int, width : int, height : int):
    img = cv2.imread(in_filename)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    
    y_0 = y
    y_1 = y + height
    x_0 = x
    x_1 = x + width
    
    crop_img = img[y_0:y_1, x_0:x_1]
    plt.imshow(cv2.cvtColor(crop_img, cv2.COLOR_BGR2RGB))
    cv2.imwrite(out_filename, crop_img)
```

```python
# there might be better ways of doing this, but the openCV function did not accept numpy? types
x = int(row["x_1"])
y = int(row["y_1"])
width = int(row["width"])
height = int(row["height"])

# WIP file structure
extract_face('./img_celeba.7z/img_celeba/000001.jpg', './cropped/000001.jpg', x, y, width, height)
```
